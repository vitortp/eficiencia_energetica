```python
# teste
def correcao_varredura_consumo(df_cpp, df_consumo):
  """
        Funcao responsavel por agrupar o consumo no turno
        Arguments:
          df_cpp: dataframe -- Dataframe com as colunas: Inicio, Fim, Turno e Produção
        Return:
          Time_i: list -- 
          Time_f: list -- 
          consumo_: list -- 
          produ__: list -- 
          tur__: list -- 
  """
  Time_i = []
  Time_f = []
  consumo_ = []
  produ__ = []
  tur__ = []
  cod_produto = []  

  for data_, turno, prod_, cod_prod in zip(df_cpp.DATA,
                                df_cpp.TURNO, df_cpp['PRODUÇÃO REAL (kg)'], df_cpp['CÓDIGO DO PRODUTO']):
    
    if turno == 'A':
      h_inicio = ' 06:00:00'
      h_fim = ' 13:59:59'
    
    elif turno == 'B':
      h_inicio = ' 14:00:00'
      h_fim = ' 21:59:59'
    
    else:
      h_inicio = ' 22:00:00'
      h_fim = ' 05:59:59'

    inicio = data_ + h_inicio
    fim = data_ + h_fim
    
    mask = (df_consumo['Time'] >= inicio) & (df_consumo['Time'] <= fim) 
    grupo = df_consumo.loc[mask]
    
    teste_2 = []
    for j in grupo.consumo:  # Substituir SEMPRE
      j = j.split(' m³')[0]
      teste_2.append(float(j))
    
    grupo.consumo = teste_2   # Substituir SEMPRE

    Time_i.append(inicio)
    Time_f.append(fim)
    consumo_.append(np.sum(grupo.consumo)) #Substituir SEMPRE
    produ__.append(prod_)
    tur__.append(turno)
    cod_produto.append(cod_prod)
    

  return Time_i, Time_f, consumo_, produ__, tur__, cod_produto
```

```python
Time_i, Time_f, consumo_, produ__, tur__, cod_produto = correcao_varredura_consumo(df_cpp, df_con)

data_ = {'Time_inicio': Time_i,
          'Time_fim': Time_f,
            'Producao_Kg': produ__,
            'consumo_Nm³': consumo_,
            'Turno': tur__,
             'cod.Produto': cod_produto}
new_df = pd.DataFrame(data_)
```

```python
def correcao_varredura_consumo(df_cpp, df_consumo):
  """
        Funcao responsavel por agrupar o consumo no turno
        Arguments:
          df_cpp: dataframe -- Dataframe com as colunas: Inicio, Fim, Turno e Produção
        Return:
          Time_i: list -- 
          Time_f: list -- 
          consumo_: list -- 
          produ__: list -- 
          tur__: list -- 
  """
  Time_i = []
  Time_f = []
  consumo_ = []
  produ__ = []
  tur__ = []

  for data_, turno, prod_ in zip(df_cpp.DATA,
                                df_cpp.TURNO, df_cpp['PRODUÇÃO REAL (kg)']):
    
    if turno == 'A':
      h_inicio = ' 06:00:00'
      h_fim = ' 13:59:59'
    
    elif turno == 'B':
      h_inicio = ' 14:00:00'
      h_fim = ' 21:59:59'
    
    else:
      h_inicio = ' 22:00:00'
      h_fim = ' 05:59:59'

    inicio = data_ + h_inicio
    fim = data_ + h_fim
    
    mask = (df_consumo['Time'] >= inicio) & (df_consumo['Time'] <= fim) 
    grupo = df_consumo.loc[mask]
    
    teste_2 = []
    for j in grupo.consumo_l10:  # Substituir SEMPRE
      j = j.split(' m³')[0]
      teste_2.append(float(j))
    
    grupo.consumo_l10 = teste_2   # Substituir SEMPRE

    Time_i.append(inicio)
    Time_f.append(fim)
    consumo_.append(np.sum(grupo.consumo_l10)) #Substituir SEMPRE
    produ__.append(prod_)
    tur__.append(turno)
    

  return Time_i, Time_f, consumo_, produ__, tur__
```
